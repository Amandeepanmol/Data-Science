```python
student={'Name':['Amandeep_Singh','Srajan_Jaiswal','Aman_Saraogi','Vibhor_Singh'],
          'Age':[20,21,19,20],
          'City':['Lucknow','Kanpur','Jhansi','Dehradun'],
          'email':['amandeep@gmail.com','jaiswalsrajan@gmail.com','aman@gmail.com','vibhu@gmail.com']
        }
```

```python
import pandas as pd
df=pd.DataFrame(student)
```

```python
df
```

```python
df['Name_City']=df['Name']+' '+df['City']            #concatenation of string
```

```python
df
```

```python
df['Name_City'].str.split(' ')                #spliting string
```

```python
df.sort_values(by='Name')                     #it is used for arraning values works both on string and integer
```

```python
df.sort_values(by='Age')
```

```python
df.sort_values(by='Name',ascending=False)                     #sorts  Name in descending order
```

```python
df.sort_values(['Name','City'],ascending=[False,True])
```
